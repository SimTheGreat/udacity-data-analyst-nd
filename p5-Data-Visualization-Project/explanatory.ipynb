```python
import pandas as pd
import numpy as np
import seaborn as sb
import matplotlib.pyplot as plt
%matplotlib inline
print('import compleate')
```

```python
df=pd.read_csv('cleaned_flight_data.csv')
```

questions:<br>

3) number of flights cancelled to airport per carrier<br>

```python
plt.figure(figsize = [12, 6])
blue=sb.color_palette()[0]
sb.countplot(data = df, y = 'carrier_name',color=blue);
plt.xticks(rotation=90)
plt.title('count of carrier_name');
```

In this visualization you can observe the most common carriers.And it is SkyWest Airlines Inc.

```python
label ='carrier', 'weather','late_aircraft','nas','secourity'
data = [df.carrier_ct.mean(),df.weather_ct.mean(),df.late_aircraft_ct.mean(), 
               df.nas_ct.mean(),df.security_ct.mean()]
plt.pie(data, labels=label, startangle=140);
plt.axis('equal')
plt.title('ct reasons');
plt.axis('equal')

plt.axis('equal');


```

this plot shows the most frucount ct with carrier ct being the most common and secourity bearly seen.Wich is interessting beacouse I expectet secourity to be one of the biggest in proportion to the others.

```python
blue=sb.color_palette()[0]
sb.barplot(data = df.sort_values(by=['carrier']), 
           y = 'carrier_name', x="arr_cancelled", color = blue)
plt.xticks(rotation=90)
plt.title('number of flights cancelled to airport per carrier');
```

here you can observe the arrived flight,carrier and the mean of these

```python
print('corrolation between delays')
sb.pairplot(df, 
             vars = ['nas_delay','late_aircraft_delay','arr_delay','nas_delay' ], 
             diag_kind = 'kde', plot_kws = {'alpha': 0.5},palette="ch:2.5,-.2,dark=.3");
```

here we can see that carrier_ct is the most common reason for a delay

```python
a=df[df.carrier_name=='SkyWest Airlines Inc.' ]
b=df[df.carrier_name=='Envoy Air' ]
c=df[df.carrier_name=='Delta Air Lines Inc.' ]
a_b_c=pd.concat([a,b,c])
a_b_c.carrier_name.value_counts()
base_color = sb.color_palette()[0]
sb.violinplot(data = a_b_c, x = 'carrier_name', y = 'arr_flights',inner=None,color=base_color)
plt.title('top 3 carriers with arriving flights')
```

here I plottet the arriving plotts in the 3 most common carriers<br>
